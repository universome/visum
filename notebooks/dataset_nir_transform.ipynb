```python
import numpy as np
import matplotlib.pyplot as plt

from utils_.visum_utils import VisumData
from utils_ import transforms as T

def get_transform(train):
    transforms = []
    # converts the image, a PIL image, into a PyTorch Tensor
    transforms.append(T.ToTensor())
    if train:
        # during training, randomly flip the training images
        # and ground-truth for data augmentation
        transforms.append(T.RandomHorizontalFlip(0.5))
        # transforms.append(T.RandomRotation(degrees=45))
    return T.Compose(transforms)

for each in  VisumData('../../../visum_data/train', modality='all', transforms=get_transform(train=True)):
    if each[0].shape[0] == 1:
        print (' - name : ', each[2], each[0].shape)
        rgb_nir = each[0].repeat(3,1,1)
        plt.imshow(np.rollaxis(rgb_nir.numpy(),0,3))
        plt.title('Name - {0}'.format(each[2]))
        plt.show()
        break
```
